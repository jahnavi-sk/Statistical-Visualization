```python
#scatter plot
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns
statisticsno



# Load the CSV file into a DataFrame
df = pd.read_csv('Dataset2_Anime.csv')

# Choose the columns for the scatter plotd
column_x = 'members'
column_y = 'episodes'

# Create a scatter plot
plt.scatter(df[column_x], df[column_y])

# Add labels and a title
plt.xlabel(column_x)
plt.ylabel(column_y)
plt.title('Scatter Plot of {} vs {}'.format(column_x, column_y))

# Show the plot
plt.show()




    
```

```python
#histogram
ratings_column = 'score'

# Create a histogram
plt.hist(df[ratings_column], bins=10, edgecolor='black')  # You can adjust the number of bins as needed

# Add labels and a title
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Anime Ratings')

# Show the plot
plt.show()




```

```python
#scatterplot matrix
from pandas.plotting import scatter_matrix

# Choose the numerical columns for the scatterplot matrix
numerical_columns = ['episodes', 'members', 'popularity', 'ranked', 'score']

# Create a scatterplot matrix
scatter_matrix(df[numerical_columns], figsize=(12, 12), diagonal='hist', alpha=0.8)

# Add labels and a title
plt.suptitle('Scatterplot Matrix of Numerical Features', y=0.95)
plt.show()




```

```python
#heatmap
import seaborn as sns

columns_of_interest = ['ranked', 'score', 'genre']

# Create a subset of the DataFrame with only the relevant columns
subset_df = df[columns_of_interest]

# Extract unique genres
unique_genres = df['genre'].explode().unique()

# Create a heatmap
heatmap_data = pd.pivot_table(subset_df, values='score', index='genre', columns='ranked', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.he-pp-atmap(heatmap_data, cmap='viridis', annot=True, fmt=".1f", linewidths=.5)

# Add labels and a title
plt.xlabel('Ranked')
plt.ylabel('Genre')
plt.title('Heatmap of Rank vs Rating within Different Categories')

# Show the plot
plt.show()
```
